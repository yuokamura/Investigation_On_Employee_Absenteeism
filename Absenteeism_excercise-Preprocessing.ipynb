# Import relevant modules

```python
import pandas as pd
```

# Import data

```python
raw_csv_data = pd.read_csv('Absenteeism-data.csv')
```

```python
# making the copy of the initial dataset
df = raw_csv_data.copy()
df
```

```python
# extract the data of the df
df.info()
```

# Data preprocessing

```python
# drop unnecesary ID table
df=df.drop(['ID'],axis = 1)
```

## introducing dummy variables

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
# lets check if we have raw with missing values
reason_columns['check']=reason_columns.sum(axis=1)
reason_columns
```

```python
reason_columns= reason_columns.drop(['check'],axis=1)
reason_columns
```

```python
# dropping the reason 0, in order to avoid the multicolinearity
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

## grouping the reason for absence (Classification)

```python
# drop the reason for absence before adding the reason_columns
df = df.drop(['Reason for Absence'],axis=1)
```

```python
# group the different reasons into four different categories.
# use .loc mehtod.  
# .max to create a new single column df 
reason_type_1=reason_columns.loc[:,1:14].max(axis=1)
reason_type_2=reason_columns.loc[:,15:17].max(axis=1)
reason_type_3=reason_columns.loc[:,18:21].max(axis=1)
reason_type_4=reason_columns.loc[:,22:].max(axis=1)
```

## concatenate columns values

```python
df=pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df
```

```python
# rename the column name of the reason for absence
# obtain the array including the names of columns
df.columns.values
```

```python
# copy and paste the array and manually rename them
columns_name = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']
```

```python
# assign new column name 
df.columns = columns_name
# check the new df
df.head()
```

## reorder columns

```python
column_names_reordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_names_reordered]
df.head()
```

# Create a checkpoint

```python
df_reason_mod = df.copy()
```

## date

```python
type(df_reason_mod['Date'][0])
```

```python
df_reason_mod['Date']=pd.to_datetime(df_reason_mod['Date'],format = '%d/%m/%Y')
df_reason_mod['Date']
```

## extract the month value

```python
## extracting the month from the first entry
df_reason_mod['Date'][0].month
```

```python
list_month=[]
for i in range(len(df_reason_mod['Date'])):
    list_month.append(df_reason_mod['Date'][i].month)
```

```python
df_reason_mod['Month Value']=list_month
df_reason_mod.head()
```

## extract the days of the week

```python
# days of the week can be checked with .weekday()
# 0 corresponds to monday, and sunday is 6
# check 
df_reason_mod['Date'][0].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

```python
# remove the date column.
df_reason_mod_no_date=df_reason_mod.drop(['Date'],axis=1)
df_reason_mod_no_date
```

```python
# rearrange the columns thus the month and days of the week appears between reason 4 and transportation expense
df_reason_mod_no_date.columns.values
```

```python
column_names = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Month Value',
       'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df_reason_mod_no_date = df_reason_mod_no_date[column_names]
df_reason_mod_no_date
```

## Checkpoint #2

```python
# we are done working with pre-processing the date columns thus we can make the checkpoint
df_reason_date_mod = df_reason_mod_no_date.copy()
```

## working with education column

```python
df_reason_date_mod['Education'].value_counts()
```

```python
# 1 corresponds to highschool graduate, 2,3,4 corresponds to graduate, postgrad, master/phd
# 1 dominates and it is sensible to put the rest into one category
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0,2:1,3:1,4:1})
```

```python
df_reason_date_mod['Education'].unique()
```

## FInal check point

```python
df_preprocessed=df_reason_date_mod.copy()
```
